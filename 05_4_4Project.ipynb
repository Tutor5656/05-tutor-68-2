```python
from datetime import timedelta, datetime

# Get user input for name and time details
user_name = input("กรุณาป้อนชื่อของคุณครับ: ")
while True:
    try:
        hours_str = input("กรุณาป้อนชั่วโมงของการทำแผลล่าสุด (0-23): ")
        hours = int(hours_str)
        if 0 <= hours <= 23:
            break
        else:
            print("ชั่วโมงไม่ถูกต้อง. กรุณาป้อนตัวเลขระหว่าง 0 ถึง 23.")
    except ValueError:
        print("ข้อมูลไม่ถูกต้อง. กรุณาป้อนตัวเลขสำหรับชั่วโมง.")

while True:
    try:
        minutes_str = input("กรุณาป้อนนาทีของการทำแผลล่าสุด (0-59): ")
        minutes = int(minutes_str)
        if 0 <= minutes <= 59:
            break
        else:
            print("นาทีไม่ถูกต้อง. กรุณาป้อนตัวเลขระหว่าง 0 ถึง 59.")
    except ValueError:
        print("ข้อมูลไม่ถูกต้อง. กรุณาป้อนตัวเลขสำหรับนาที.")

while True:
    try:
        notification_interval_hours_str = input("กรุณาป้อนช่วงเวลาแจ้งเตือน (เป็นชั่วโมง): ")
        notification_interval_hours = int(notification_interval_hours_str)
        if notification_interval_hours > 0:
            break
        else:
            print("ช่วงเวลาแจ้งเตือนต้องมากกว่า 0.")
    except ValueError:
        print("ข้อมูลไม่ถูกต้อง. กรุณาป้อนตัวเลขสำหรับช่วงเวลาแจ้งเตือน.")

# Define necessary variables (current_datetime will be used for the date part)
current_datetime = datetime.now()

# Create a datetime object for the last wound dressing time
# Using the date from current_datetime and time from hours, minutes
last_dressing_time_dt = current_datetime.replace(hour=hours, minute=minutes, second=0, microsecond=0)

# Create a timedelta object from the notification interval
notification_timedelta = timedelta(hours=notification_interval_hours)

# Calculate the next wound dressing time
next_dressing_time_dt = last_dressing_time_dt + notification_timedelta

print(f"\nสวัสดีคุณ {user_name}!")
print(f"เวลาทำแผลล่าสุด: {last_dressing_time_dt.strftime('%Y-%m-%d %H:%M')}")
print(f"ช่วงเวลาแจ้งเตือน: {notification_interval_hours} ชั่วโมง")
print(f"เวลาทำแผลครั้งถัดไป: {next_dressing_time_dt.strftime('%Y-%m-%d %H:%M')}")
```
